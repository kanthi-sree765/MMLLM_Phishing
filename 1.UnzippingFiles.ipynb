```python
import os
import zipfile

data_dir = r'C:\Users\DELL\Desktop\Phishing\data'

for root, dirs, files in os.walk(data_dir):
    for file in files:
        if file.endswith('.zip'):
            zip_path = os.path.join(root, file)
            extract_path = os.path.splitext(zip_path)[0]  # remove .zip extension
            if not os.path.exists(extract_path):  # avoid re-extracting
                with zipfile.ZipFile(zip_path, 'r') as zip_ref:
                    zip_ref.extractall(extract_path)
                    print(f'Extracted: {zip_path} → {extract_path}')
```

```python
import os

data_dir = r'C:\Users\DELL\Desktop\Phishing\data\MMLLM_Phishing\c'

count = 0  # to count how many files were found

for root, dirs, files in os.walk(data_dir):
    for file in files:
        if file.endswith('.png') or file.endswith('.html'):
            file_path = os.path.join(root, file)
            with open(file_path, 'rb') as f:
                content = f.read()
                print(f"[+] Found file: {file_path}")
                print(f"    - Size: {len(content)} bytes")
            count += 1

print(f"\n✅ Total phishing files found: {count}")
```
